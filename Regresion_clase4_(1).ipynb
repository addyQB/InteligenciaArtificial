# Metricas de evaluacion

En el caso de los modelos de regresion, tenemos que predecir valores continuos y no clases de manera individual, por esto, este tipo de modelo se evaluan distinto de como evaluariamos el rendimiento de un modelo de clasificacion. A continuacion presentaremos las mas populares.

- MAE (Mean absolute error)
- MSE (Mean squared error)
- RMSE (Root mean squared error)
- R squared

## Error Cuadrático Medio (MSE):

**Definición:**

 El Error Cuadrático Medio es una medida del promedio de las diferencias al cuadrado entre los valores predichos y los valores reales en el conjunto de datos.

**Interpretación:**

 Nos da una idea de qué tan lejos, en promedio, están los valores predichos de los valores reales. Un MSE más pequeño indica un mejor desempeño del modelo, siendo valores cercanos a cero los ideales.

**Fórmula:**

\begin{align}
       MSE= \frac{1}{N} \sum_{i=1}^{N} (y_{real} - y_{pred})^2
\end{align}




 donde y_real son los valores reales del objetivo, y_pred son los valores predichos del objetivo, y n es el número de puntos de datos.


## Error absoluto medio (MAE):

**Definición:**

 El Error Absoluto Medio es una medida del promedio de las diferencias absolutas entre los valores predichos y los valores reales en el conjunto de datos.

**Interpretación:**

 El MAE representa la magnitud promedio de los errores cometidos por el modelo, sin considerar su dirección. Es menos sensible a valores atípicos en comparación con el MSE.

**Fórmula:**

\begin{align}
       MAE= \frac{1}{N} \sum_{i=1}^{N} |y_{real} - y_{pred}|
\end{align}



## Root mean squared error (RMSE)

**Definición:**

 El RMSE es la raíz cuadrada del MSE. Se utiliza para proporcionar una medida más interpretable del rendimiento del modelo en las unidades originales de la variable objetivo.


**Interpretación:**

Al igual que el MSE, un RMSE más pequeño indica un mejor desempeño del modelo, y los valores cercanos a cero son preferidos.


**Formula:**
\begin{align}
       RMSE= \sqrt{MSE}
\end{align}

Tomen en cuenta que esta metrica es sensible a los outliers.



NOTA:

Un valor de error absoluto medio perfecto es 0,0, lo que significa que todas las predicciones coincidieron exactamente con los valores esperados.

Este casi nunca es el caso, y si sucede, sugiere que su problema de modelado predictivo es trivial.

Una buena metrica de evaluacion es relativa a su conjunto de datos específico.

Es una buena idea establecer primero una metrica de referencia para su conjunto de datos utilizando un modelo predictivo ingenuo, como predecir el valor objetivo medio del conjunto de datos de entrenamiento.

## R Squared

**Definición:**

 El R-cuadrado es una medida estadística que representa la proporción de la varianza en la variable objetivo que es explicada por el modelo. Mide qué tan bien el modelo se ajusta a los datos en comparación con un promedio simple de la variable objetivo.


**Interpretación:**

 El R-cuadrado varía de 0 a 1. Un valor de 1 indica que el modelo se ajusta perfectamente a los datos, mientras que un valor de 0 significa que el modelo no explica la varianza en los datos mejor que un promedio simple. Los valores negativos de R2 indican un ajuste deficiente del modelo.


**Formula:**

\begin{align}
 R^2 =  1 - \frac{\sum_i (y_{real} - y_{pred})^2}{\sum_i (y_{real} - y_{promedio})^2}
\end{align}

donde y_promedio es el promedio de los valores reales del objetivo.




Como podemos ver arriba, en el numerador de la fórmula, restamos los valores predichos de nuestro modelo de los valores reales y los elevamos al cuadrado. En el denominador, quitamos el valor promedio de nuestro valor dependiente de los valores fundamentales y lo elevamos al cuadrado.

En resumen, estas métricas de evaluación ayudan a evaluar el desempeño de modelos de regresión. Valores más bajos de MSE, RMSE y MAE son deseables, ya que indican un ajuste más cercano a los valores reales. Para R-cuadrado, valores más altos cercanos a 1 indican un mejor ajuste del modelo, mientras que valores negativos sugieren que el modelo es peor que un promedio simple. Evaluar múltiples métricas proporciona una comprensión completa de la efectividad del modelo en hacer predicciones.

## Regresión Lineal

La regresión lineal es un algoritmo de aprendizaje supervisado utilizado para predecir una variable continua (valor numérico) basándose en la relación lineal entre una o más variables predictoras (también conocidas como características o variables independientes). En su forma más simple, la regresión lineal se representa mediante una línea recta que mejor se ajusta a los datos para hacer predicciones.

**Ventajas de la Regresión Lineal:**

- **Simplicidad:**

 La regresión lineal es fácil de entender e implementar. Su simplicidad la hace un buen punto de partida para problemas de predicción sencillos.

- **Interpretación:**

 Las relaciones lineales son fáciles de interpretar. Los coeficientes de la regresión representan la dirección y la magnitud del efecto de cada variable en la variable objetivo.

- **Eficiencia computacional:**

 La regresión lineal es computacionalmente eficiente, lo que la hace adecuada para conjuntos de datos grandes.

**Desventajas de la Regresión Lineal:**

- **Linealidad requerida:**

 La regresión lineal asume que la relación entre las variables predictoras y la variable objetivo es lineal. Si la relación es no lineal, la regresión lineal puede no ajustarse bien a los datos.

- **Sensible a valores atípicos:**

 La regresión lineal es sensible a la presencia de valores atípicos (outliers), que pueden afectar significativamente los resultados.

- **Multicolinealidad:**

 Si hay una alta correlación entre las variables predictoras (multicolinealidad), la regresión lineal puede tener problemas de inestabilidad y no proporcionar estimaciones precisas de los coeficientes.

- **Sobreajuste o subajuste:**

 Si el modelo es demasiado complejo o muy simple, puede llevar al sobreajuste (overfitting) o subajuste (underfitting) respectivamente, lo que afecta la calidad de las predicciones.



**Suposiciones de la Regresión Lineal:**

Para obtener resultados confiables y precisos con la regresión lineal, es importante que los datos cumplan con ciertas suposiciones:

- **Linealidad:**

 La relación entre las variables predictoras y la variable objetivo debe ser lineal. Se puede verificar gráficamente utilizando gráficos de dispersión.

- **Independencia de errores:**

 Los errores de predicción deben ser independientes y no correlacionados entre sí. Los residuos deben estar dispersos alrededor de cero en un gráfico de residuos vs. valores predichos.

- **Homocedasticidad:**

 La varianza de los errores de predicción debe ser constante a lo largo de todas las posibles combinaciones de valores de las variables predictoras. Se puede verificar mediante un gráfico de residuos vs. valores predichos o mediante pruebas estadísticas.



image.png

- **No multicolinealidad:**

 Las variables predictoras deben ser independientes y no estar altamente correlacionadas entre sí para evitar problemas de inestabilidad y sesgo en los coeficientes.

- **Normalidad de errores:**

 Los errores de predicción deben seguir una distribución normal. Se puede verificar mediante un gráfico de cuantiles normales o pruebas estadísticas como el test de Shapiro-Wilk.



image.png

Cuando estas suposiciones se cumplen, la regresión lineal puede proporcionar estimaciones precisas y significativas para hacer predicciones y obtener información valiosa sobre las relaciones entre las variables en el modelo. Sin embargo, es esencial evaluar y validar estas suposiciones antes de usar la regresión lineal para tomar decisiones basadas en los resultados.

# Ridge regression

La regresión ridge es un método de ajuste de modelo que se utiliza para analizar cualquier dato que sufra de multicolinealidad. Este método realiza la regularización L2. Cuando ocurre el problema de la multicolinealidad, los mínimos cuadrados no están sesgados y las varianzas son grandes, lo que da como resultado que los valores predichos estén muy lejos de los valores reales.

Este tipo de modelos se usan para reducir los parametros, por lo que previene la multi colinearidad. Reduce la complejidad del modelo al disminuir el coeficiente.


Los supuestos de la regresión ridge son los mismos que los de la regresión lineal: linealidad, varianza constante e independencia. Sin embargo, dado que la regresión ridge no proporciona límites de confianza, no es necesario suponer que la distribución de errores sea normal.

## Stepwise regression

En cualquier fenómeno, habrá ciertos factores que jugarán un papel más importante en la determinación de un resultado. En términos simples, la regresión por pasos es un proceso que ayuda a determinar qué factores son importantes y cuáles no. Ciertas variables tienen un valor p bastante alto y no contribuyeron significativamente a la precisión de nuestra predicción. A partir de ahí, solo se mantienen los factores importantes para garantizar que el modelo lineal haga su predicción en función de los factores que pueden ayudarlo a producir el resultado más preciso.
