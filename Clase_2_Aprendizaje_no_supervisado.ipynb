# Aprendizaje no supervisado

## Reducción de dimensionalidad

La reducción de dimensionalidad juega un papel muy importante en el area de aprendizaje no supervisado por algunas razones:

**Visualización**

Los datos de alta dimensionalidad pueden ser difíciles de visualizar y comprender. Técnicas de reducción de dimensionalidad como Análisis de Componentes Principales (PCA) y Embedding Estocástico Vecino t-Distribuido (t-SNE) transforman los datos a un espacio de menor dimensionalidad, a menudo dos o tres dimensiones, lo que facilita su visualización y exploración.

Reduccion de dimensiones1.png

**Eficiencia computacional**

Los datos de alta dimensionalidad pueden llevar a una mayor complejidad computacional y uso de memoria. La reducción de la dimensionalidad permite que los algoritmos sean más eficientes computacionalmente, lo que hace posible manejar grandes conjuntos de datos y acelerar los cálculos.

**Feature engineering** o **Ingeniería de características**

En muchos casos, los conjuntos de datos pueden contener características redundantes o irrelevantes. La reducción de dimensionalidad ayuda a seleccionar las características más informativas y descartar ruido o información redundante. Este proceso de ingeniería de características mejora el rendimiento y la generalización del modelo.

PCA1.png

**Maldicion de la dimensionalidad**

La "Maldición de la Dimensionalidad" se refiere al problema de la mayor dispersión de datos y densidad de muestras a medida que el número de dimensiones aumenta. A medida que aumenta la dimensionalidad, los puntos de datos se dispersan más, y el riesgo de sobreajuste aumenta. La reducción de dimensionalidad puede mitigar este problema al reducir el número de dimensiones.

**Mejora el resultado del agrupamiento**

La reducción de dimensionalidad puede mejorar los resultados de agrupamiento (clustering) al desentrañar los puntos de datos y reducir el impacto de características irrelevantes. Esto permite a los algoritmos de agrupamiento descubrir patrones y estructuras más significativos en los datos.

**Reducción de ruido**

Los datos de alta dimensionalidad pueden contener ruido y errores de medición. Al reducir la dimensionalidad, podemos centrarnos en los patrones más significativos y reducir la influencia del ruido, lo que conduce a modelos de aprendizaje no supervisado más robustos y precisos.

**Mejor generalización**

Los datos de alta dimensionalidad pueden provocar sobreajuste en los modelos de aprendizaje no supervisado. La reducción de dimensionalidad puede ayudar a crear representaciones más compactas y generalizadas de los datos, lo que resulta en un mejor rendimiento de generalización.

Dentro del area de los algoritmos que podemos utilizar para la reducción de dimensiones, tenemos dos ramas principales:




### Proyección lineal

"*Las técnicas de proyección lineal buscan encontrar un subespacio lineal en el espacio de alta dimensionalidad donde los datos puedan ser representados de manera efectiva con dimensiones más bajas. Mapean los puntos de datos a un espacio de menor dimensionalidad mediante transformaciones lineales.*"

Las técnicas de proyección lineal son métodos que se utilizan para hacer que un problema complejo sea más fácil de entender y trabajar. Imagina que tienes un montón de datos grandes y desordenados, y quieres organizarlos y simplificarlos.

Cuando decimos "representado de manera efectiva con menores dimensiones", nos referimos a tomar todos los datos desordenados en el espacio de alta dimensión y encontrar una forma más simple de mostrarlos en un espacio más pequeño, como un pedazo de papel (2D) o una línea (1D).

**¿Que es un espacio de alta dimension?**

Piensa en un espacio de alta dimensión como un mundo con muchas dimensiones o direcciones. Por ejemplo, nuestro mundo cotidiano tiene tres dimensiones (longitud, anchura y altura). Pero los datos pueden tener muchas más dimensiones, como temperatura, tiempo, brillo, etc.

**Ejemplos de algoritmos de proyección lineal:**


- **Análisis de Componentes Principales (PCA):**

 Una técnica lineal conocida que encuentra los componentes principales (ejes ortogonales) para proyectar los datos mientras maximiza la varianza a lo largo de cada componente.
 PCA (Análisis de Componentes Principales) logra esto abordando la correlación entre características. Si la correlación es muy alta entre un subconjunto de características, PCA intentará combinar las características altamente correlacionadas y representar estos datos con un número menor de características linealmente no correlacionadas. El algoritmo sigue realizando esta reducción de correlación, encontrando las direcciones de máxima varianza en los datos originales de alta dimensión y proyectándolos en un espacio dimensional más pequeño. Estos componentes recién derivados se conocen como componentes principales.

 PCA no utiliza etiquetas de clase durante su cálculo, por lo que es una técnica no supervisada. Se enfoca solo en la estructura de los datos y las relaciones entre las características.

- **Análisis Discriminante Lineal (LDA):**

 Similar a PCA, pero también considera las etiquetas de clase para encontrar un subespacio lineal que maximice la separación entre clases. LDA también reduce el número de características (dimensiones) en los datos, pero su objetivo principal es mejorar la capacidad de los datos para discriminar entre diferentes clases. Busca nuevas direcciones en los datos que maximicen la separación entre diferentes clases y minimicen la variación dentro de cada clase. Estas direcciones se llaman discriminantes. A diferencia de PCA, LDA busca preservar la separabilidad entre clases, no solo la varianza total. Se enfoca en maximizar las diferencias entre las clases.

 LDA requiere etiquetas de clase durante su cálculo, por lo que es una técnica supervisada. Considera tanto la estructura de los datos como la información sobre las clases.


En pocas palabras, PCA se enfoca más en reducir la complejidad y visualizar datos, mientras que LDA se centra en mejorar la separación de clases y preparar datos para tareas de clasificación.



#### Ejemplo de PCA

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
```

```python
#Cargamos el dataset de Iris con el que trabajamos la clase pasada
iris = load_iris()
X, y = iris.data, iris.target

# Tomamos la libreria de PCA de Sklearn y la utilizamos para reduccion de caracteristicas
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python
# Imprimimos los datos originales en 2D
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', edgecolors='k')
plt.title('Original Data (2D)')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.show()
```

```python
#PCA
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', edgecolors='k')
plt.title('PCA: Dimensionality Reduction to 2D')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

#### Diferencia entre PCA y LDA:


```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

lda = LinearDiscriminantAnalysis(n_components=2)
X_lda = lda.fit_transform(X, y)

# Plot the results of PCA and LDA
plt.figure(figsize=(10, 4))

# Plot PCA results
plt.subplot(1, 2, 1)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', edgecolors='k')
plt.title('PCA: Dimensionality Reduction to 2D')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')

# Plot LDA results
plt.subplot(1, 2, 2)
plt.scatter(X_lda[:, 0], X_lda[:, 1], c=y, cmap='viridis', edgecolors='k')
plt.title('LDA: Dimensionality Reduction to 2D')
plt.xlabel('Linear Discriminant 1')
plt.ylabel('Linear Discriminant 2')

plt.tight_layout()
plt.show()
```

Evaluación de los resultados de PCA y LDA:

```python

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Perform PCA for dimensionality reduction to 2D
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

# Perform LDA for dimensionality reduction to 2D
lda = LinearDiscriminantAnalysis(n_components=2)
X_train_lda = lda.fit_transform(X_train, y_train)
X_test_lda = lda.transform(X_test)

# Create a Logistic Regression classifier
classifier = LogisticRegression()

# Fit the classifier on PCA-transformed data and make predictions
classifier.fit(X_train_pca, y_train)
y_pred_pca = classifier.predict(X_test_pca)

# Fit the classifier on LDA-transformed data and make predictions
classifier.fit(X_train_lda, y_train)
y_pred_lda = classifier.predict(X_test_lda)

# Evaluate the results for PCA
print("Results for PCA:")
print("Accuracy:", accuracy_score(y_test, y_pred_pca))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_pca))
print("Classification Report:\n", classification_report(y_test, y_pred_pca))

```

```python

# Evaluate the results for LDA
print("\nResults for LDA:")
print("Accuracy:", accuracy_score(y_test, y_pred_lda))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_lda))
print("Classification Report:\n", classification_report(y_test, y_pred_lda))
```

#### Analisis de correspondencia

El Análisis de Correspondencias (CA, por sus siglas en inglés) es un método estadístico utilizado para la reducción de dimensionalidad y visualización de datos categóricos. Es similar al Análisis de Componentes Principales (PCA), pero está diseñado específicamente para analizar las relaciones entre variables categóricas en lugar de variables continuas.

El objetivo principal del Análisis de Correspondencias es explorar y descubrir las asociaciones entre filas y columnas en una tabla de contingencia. Una tabla de contingencia es una tabla que muestra la distribución de frecuencias de dos o más variables categóricas. Muestra cómo se distribuyen las categorías de una variable a lo largo de las categorías de otra variable.

Algunos otros ejemplos de PCA:

- Incremental PCA:

El PCA es potente para reducir las dimensiones de un conjunto de datos manteniendo la mayor parte de su variabilidad. Sin embargo, puede ser computacionalmente costoso para conjuntos de datos grandes. El PCA Incremental aborda esta limitación procesando los datos en pequeños lotes (trozos) en lugar de cargar todo el conjunto de datos en memoria de una vez. Esto lo hace más eficiente en cuanto al uso de memoria y permite procesar conjuntos de datos grandes que podrían no caber completamente en la memoria.


- Sparse PCA:

En el PCA Escaso, las componentes principales son escasas, lo que significa que tienen principalmente valores cero. Esta propiedad de esparcimient o ayuda a identificar y seleccionar un pequeño subconjunto de las características más relevantes, lo que proporciona una representación más interpretable y eficiente de los datos.

- Kernel PCA:

Mientras que el PCA tradicional funciona mejor para datos linealmente separables, el PCA con Núcleo puede manejar patrones complejos y relaciones no lineales. Logra esto al mapear los datos a un espacio de mayor dimensión utilizando una función no lineal (función núcleo) y luego aplicar el PCA para encontrar las componentes principales en este espacio transformado. El resultado es una representación de menor dimensión de los datos que captura estructuras no lineales y puede ser útil para varias tareas de aprendizaje automático.

### Mainfold learning

En términos más sencillos, imagina un objeto tridimensional como una dona. Aunque un dona existe en tres dimensiones (x, y y z), se puede representar en un plano bidimensional (x, y) sin perder mucha información sobre su forma. De manera similar, el aprendizaje de variedades busca encontrar una representación de menor dimensión de los puntos de datos mientras preserva las estructuras importantes y las relaciones entre ellos.

- **t-SNE (t-distributed Stochastic Neighbor Embedding):**

t-SNE se utiliza comúnmente para visualización porque puede capturar eficazmente las relaciones locales y revelar la estructura de agrupación de los datos.
El algoritmo funciona midiendo las similitudes entre pares de puntos de datos en el espacio de alta dimensión y el espacio de menor dimensión.
Trata de preservar las similitudes entre puntos cercanos mientras exagera las diferencias entre puntos distantes.
Como resultado, t-SNE es particularmente útil para visualizar grupos o conjuntos de puntos de datos en un gráfico 2D o 3D.


- **Isomap (Isometric Mapping):**

Construye un grafo que conecta los puntos de datos en función de sus distancias mutuas y luego utiliza teoría de grafos para encontrar la ruta más corta (distancia geodésica) entre cualquier par de puntos de datos.
De esta manera, Isomap puede capturar eficazmente la estructura global de la variedad subyacente, incluso en casos en los que t-SNE podría tener dificultades.
Isomap puede ser útil para tareas como reducción de dimensionalidad no lineal y aprendizaje de variedades cuando los puntos de datos no están bien agrupados.

## Clustering o agrupamiento

Como mencionamos en la presentación, clustering o agrupamiento es una de las herramientas mas conocidas en el area de aprendizaje no supervisado, ya que intenta encontrar relaciones entre los datos. Si estos datos no estan etiquetados, nos permite saber un poco mas de las caracteristicas de los datos que nos permitan generar despues un modelo.

Para datos etiquetados, los algoritmos de clustering tambien se pueden utilizar para detectar outliers, o para descubrir patrones dentro de la clasificacion de las etiquetas.

### Agrupamiento k-medias

K-Means es un popular algoritmo de aprendizaje automático no supervisado que se utiliza para agrupar datos. Su principal objetivo es particionar un conjunto de datos dado en K grupos (clusters), donde cada punto de datos pertenece al grupo cuyo centroide (media) está más cercano. El "media" en K-Means se refiere al centro de cada cluster.

Si nosotros descibieramos los pasos como un pseudocodigo, lo podriamos describir de la siguiente manera:

- **Paso Inicial:**

 Selecciona K puntos de datos aleatorios del conjunto de datos como centroides iniciales. Estos puntos servirán como centros de los clusters iniciales.

- **Paso de Asignación:**

 Para cada punto de datos en el conjunto, calcula la distancia a cada centroide y asigna el punto de datos al cluster cuyo centroide esté más cercano.

- **Paso de Actualización:**

 Después de que todos los puntos de datos se hayan asignado a clusters, calcula los nuevos centroides de cada cluster tomando la media de todos los puntos de datos en ese cluster.

- **Repetir:**

 Los pasos 2 y 3 se repiten de forma iterativa hasta que los centroides convergen, es decir, ya no cambian significativamente o se alcanza el número máximo de iteraciones.

- **Convergencia:**

 El algoritmo se detiene cuando los centroides ya no cambian o se alcanza el número máximo de iteraciones. En este punto, se considera que los clusters son estables.

```python
import numpy as np

def kmeans(X, K, max_iters=10):
    np.random.seed(42)
    centroids = X[np.random.choice(X.shape[0], K, replace=False)]

    for iter in range(max_iters):
        # Asignamos los datos a los puntos de centroide mas cercano
        labels = np.argmin(np.linalg.norm(X[:, np.newaxis] - centroids, axis=2), axis=1)
         #Actualizamos los centroides promediando los puntos para cada cluster
        new_centroids = np.array([X[labels == k].mean(axis=0) for k in range(K)])

        print(f"Iteration {iter+1}:")
        print("Centroids:")
        print(new_centroids)
        print("Cluster labels:")
        print(labels)
        print("=" * 30)

        # Revisamos la convergencia
        if np.all(centroids == new_centroids):
            break
        centroids = new_centroids
    return centroids, labels
```

### ¿Como podemos eleguir el criterio de numero de clusters?

Haciendo contraste con herramientas de aprendizaje supervisado, en donde tenemos problemas de clasificacion o regresion, el agrupamiento requiere mas esfuerzo para elegir el criterio de optimizacion. Usualmente, cuando se trabaja con kmedias, debemos de optimizar la suma de las distancias cuadradas entre las observaciones o puntos y sus centroides. Para esto tenemos la siguiente formula:

\begin{align}
        J(C) = \sum_{k=1}^K \sum_{i \in C_k}|| x_i - \mu_k || \rightarrow \min_C
\end{align}

En donde interpretamos a C como el conjunto de clusters a la potencia k. $\mu_k$ es el centroide de un cluster $C_k$.

Esta definicion pareceria estar correcta, sin embargo, sin restriccion, el optimo se alcanza cuando el numero de centroides sea igual al numero de observaciones, por lo que cada uno de los puntos terminaria con su propio cluster. Para evitar sto, tenemos la funcion:

\begin{align}
        D(k) = \frac{|J(C_k) - J(C_{k+1})|}{|J(C_{k-1} - J(C_k))|} \rightarrow \min_k
\end{align}

```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

X = np.zeros((150, 2))

np.random.seed(seed=42)
X[:50, 0] = np.random.normal(loc=0.0, scale=.3, size=50)
X[:50, 1] = np.random.normal(loc=0.0, scale=.3, size=50)

X[50:100, 0] = np.random.normal(loc=2.0, scale=.5, size=50)
X[50:100, 1] = np.random.normal(loc=-1.0, scale=.2, size=50)

X[100:150, 0] = np.random.normal(loc=-1.0, scale=.2, size=50)
X[100:150, 1] = np.random.normal(loc=2.0, scale=.5, size=50)

plt.figure(figsize=(5, 5))
plt.plot(X[:, 0], X[:, 1], 'bo');
```

```python
from sklearn.cluster import KMeans

inertia = []
for k in range(1, 8):
    kmeans = KMeans(n_clusters=k, random_state=1).fit(X)
    inertia.append(np.sqrt(kmeans.inertia_))
```

```python
plt.plot(range(1, 8), inertia, marker='s');
plt.xlabel('$k$')
plt.ylabel('$J(C_k)$')
```

Con esta funcion vemos que $J(C_k)$ llega hasta el punto 3, en donde a partir de ahi no tiene cambios grandes, por lo que podemos inferir que ese es el valor optimo de clusters.

Este tipo de problema es np-dificil, ya que su complejidad va en aumento dependiento de las dimensiones, los clusters y el numero de observaciones. La implementacion de este tipo de problema con sklearn tiene la ventaja de su funcion de inicializacion default nos ayuda a identificar centroides robustos. El StandardScaler es una técnica de preprocesamiento comúnmente utilizada en el aprendizaje automático para estandarizar o normalizar las características de un conjunto de datos. Garantiza que todas las características tengan la misma escala, lo cual es importante para que muchos algoritmos de aprendizaje automático funcionen de manera efectiva.


La aplicación del StandardScaler en el algoritmo de K-Means es importante para garantizar que todas las características tengan la misma escala. Esto es necesario porque K-Means utiliza la distancia euclidiana para medir la similitud entre puntos de datos y los centroides de los clústeres.

Si las características tienen diferentes escalas, algunas de ellas podrían dominar más en el cálculo de la distancia, llevando a resultados sesgados y afectando el proceso de agrupación.
